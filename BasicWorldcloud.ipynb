# Worldcloud visualiztion of my statement of purpose

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = open("./corpus/sop.txt").read()

stopwords = None

wc = WordCloud(background_color="white",
               max_words = 100,          # max words on picture
               stopwords=stopwords)      # None will use default stop words
wc.generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
